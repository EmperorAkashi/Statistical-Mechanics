```python
from measurement import *
from matplotlib import rcParams
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
%matplotlib inline
```

```python
relaxation = [500] * 50
```

```python
generate_equilibration_plots("./ising_10", "E")
```

```python
data = MCDataFrame("./ising_5", relaxation_times=relaxation)
df = data.df
df = df.reset_index(drop=True, inplace=True)
data2 = MCDataFrame("./ising_10", relaxation_times=relaxation)
df2 = data2.df
df2 = df2.reset_index(drop=True, inplace=True)
```

```python
fig, ax = plt.subplots()
ax.errorbar(df['T'], df['C'], yerr=df['Cerr'], 
            fmt='k:', elinewidth=1, capsize=2, label='$L = 5$', 
            markerfacecolor='none', markersize=4)
max_5 = df['C'].idxmax()
print(max_5)
max_5 = df.iloc[max_5]
ax.errorbar(max_5['T'], max_5['C'], yerr=max_5['Cerr'],
            fmt='rs:', elinewidth=1, capsize=2, 
            markerfacecolor='none', markersize=4)
ax.errorbar(df2['T'], df2['C'], yerr=df2['Cerr'],
            fmt='k:', elinewidth=1, capsize=2, label='$L = 10$', 
            markerfacecolor='none', markersize=4)
max_10 = df2['C'].idxmax()
max_10 = df2.iloc[max_10]
ax.errorbar(max_10['T'], max_10['C'], yerr=max_10['Cerr'],
            fmt='rs:', elinewidth=1, capsize=2, 
            markerfacecolor='none', markersize=4)
ax.grid(linestyle='--')
ax.legend()
ax.set_ylim(0, 3)
```

```python
from measurement import *
from matplotlib import rcParams
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
%matplotlib inline
```

```python
relaxation = [500] * 40
```

```python
data_5 = MCDataFrame("./data_ising/ising_5", relaxation_times=relaxation)
data_10 = MCDataFrame("./data_ising/ising_10", relaxation_times=relaxation)
data_15 = MCDataFrame("./data_ising/ising_15", relaxation_times=relaxation)
data_20 = MCDataFrame("./data_ising/ising_20", relaxation_times=relaxation)
data_25 = MCDataFrame("./data_ising/ising_25", relaxation_times=relaxation)
data_30 = MCDataFrame("./data_ising/ising_30", relaxation_times=relaxation)
```

```python
name = 'E'
name_err = name + 'err'
fig, ax = plt.subplots()
#ax.set_yscale('log')
ax.errorbar(data_5.df['T'], data_5.df[name], data_5.df[name_err],
            fmt='ks:', elinewidth=1, capsize=2, label='$L = 5$', 
            markerfacecolor='none', markersize=4)
ax.errorbar(data_10.df['T'], data_10.df[name], data_10.df[name_err],
            fmt='bs:', elinewidth=1, capsize=2, label='$L = 10$',
            markerfacecolor='none', markersize=4)
ax.errorbar(data_15.df['T'], data_15.df[name], data_15.df[name_err],
            fmt='rs:', elinewidth=1, capsize=2, label='$L = 15$', 
            markerfacecolor='none', markersize=4)
ax.errorbar(data_20.df['T'], data_20.df[name], data_20.df[name_err],
            fmt='gs:', elinewidth=1, capsize=2, label='$L = 20$', 
            markerfacecolor='none', markersize=4)
ax.errorbar(data_25.df['T'], data_25.df[name], data_25.df[name_err],
            fmt='ys:', elinewidth=1, capsize=2, label='$L = 25$', 
            markerfacecolor='none', markersize=4)
ax.errorbar(data_30.df['T'], data_30.df[name], data_30.df[name_err],
            fmt='ms:', elinewidth=1, capsize=2, label='$L = 30$', 
            markerfacecolor='none', markersize=4)
ax.grid(linestyle='--')
ylabel = "${0}$".format(name)
ax.set_ylabel(ylabel, fontsize=14)
ax.set_xlabel('$T$', fontsize=14)
ax.set_title('Free Energy of the 3D Ising Model', y=1.03)
ax.legend(loc='best')
filename = "{0}_ising.pdf".format(name)
fig.savefig(filename)
```
